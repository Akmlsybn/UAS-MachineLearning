**Data Split**

```python
from pathlib import Path
import shutil
import random

from sklearn.model_selection import train_test_split

print("Library berhasil dimuat")
```

```python
RANDOM_STATE = 42

random.seed(RANDOM_STATE)

print("Random seed:", RANDOM_STATE)
```

```python
DATASET_PATH = Path(
    "/content/drive/MyDrive/Dataset ML/Dataset_256_Preprocessing"
)

print(DATASET_PATH.exists())
```

```python
SPLIT_PATH = Path(
    "/content/drive/MyDrive/Dataset ML/Dataset_Split"
)

SPLIT_PATH.mkdir(
    parents=True,
    exist_ok=True
)

print(SPLIT_PATH)
```

```python
classes = ["Slingbag", "Totebag"]

for split in ["train", "validation", "test"]:

    for class_name in classes:

        (
            SPLIT_PATH /
            split /
            class_name
        ).mkdir(
            parents=True,
            exist_ok=True
        )

print("Struktur folder berhasil dibuat")
```

```python
for class_name in classes:

    folder = DATASET_PATH / class_name

    images = list(folder.glob("*.jpg"))

    # 70% train
    train_images, temp_images = train_test_split(
        images,
        test_size=0.30,
        random_state=RANDOM_STATE
    )

    # 15% validation
    # 15% test
    validation_images, test_images = train_test_split(
        temp_images,
        test_size=0.50,
        random_state=RANDOM_STATE
    )

    print(f"\n{class_name}")
    print(f"Train      : {len(train_images)}")
    print(f"Validation : {len(validation_images)}")
    print(f"Test       : {len(test_images)}")
```

```python
for class_name in classes:

    folder = DATASET_PATH / class_name

    images = list(folder.glob("*.jpg"))

    train_images, temp_images = train_test_split(
        images,
        test_size=0.30,
        random_state=RANDOM_STATE
    )

    validation_images, test_images = train_test_split(
        temp_images,
        test_size=0.50,
        random_state=RANDOM_STATE
    )

    for image_path in train_images:

        shutil.copy(
            image_path,
            SPLIT_PATH /
            "train" /
            class_name /
            image_path.name
        )

    for image_path in validation_images:

        shutil.copy(
            image_path,
            SPLIT_PATH /
            "validation" /
            class_name /
            image_path.name
        )

    for image_path in test_images:

        shutil.copy(
            image_path,
            SPLIT_PATH /
            "test" /
            class_name /
            image_path.name
        )

print("Proses split selesai")
```

```python
for split in ["train", "validation", "test"]:

    print(f"\n=== {split.upper()} ===")

    total = 0

    for class_name in classes:

        count = len(
            list(
                (
                    SPLIT_PATH /
                    split /
                    class_name
                ).glob("*.jpg")
            )
        )

        total += count

        print(f"{class_name}: {count}")

    print(f"Total: {total}")
```
