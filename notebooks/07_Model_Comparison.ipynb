**Model Comparison**

```python
import pandas as pd
import matplotlib.pyplot as plt

results = pd.DataFrame({
    "Model": [
        "HOG+SVM",
        "EfficientNetB0",
        "MobileNetV2"
    ],
    "Accuracy": [
        92.00,
        99.33,
        100.00
    ]
})

results
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    results["Model"],
    results["Accuracy"]
)

plt.ylabel("Accuracy (%)")

plt.title(
    "Perbandingan Akurasi Model"
)

plt.ylim(85,101)

plt.show()
```
